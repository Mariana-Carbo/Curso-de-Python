````python
def comparar(serie_1,serie_2):
    if serie_1 == serie_2:
      return 'verdadero'
    elif serie_1 != serie_2:
      return ('falso')
````

````python
def comparar(serie_1,serie_2):
  for s1, s2 in zip(serie_1,serie_2):
    if s1 == s2:
      print('verdadero')
    elif s1 != s2:
      print ('falso')
````
